```python
#filter the dataset to include only the years 2009 to 2019
df_filtered = df[(df['year'] >= 2009) & (df['year'] <= 2019)]
df_filtered.head()
```

```python
# Get unique years
unique_years = df['year'].unique()

# Create a dictionary to store each year's DataFrame
dfs_by_year = {}

for year in unique_years:
    # Filter df for the specific year
    dfs_by_year[year] = df[df['year'] == year]
#list of years
list(dfs_by_year.keys())
```
